```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import math
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import IsolationForest
```

```python
receiver = ['1','2','3']
location_vertical = ['1','2','3','4']
location_horizontal = ['1','2','3','4','5']
loc_list = []
for horizontal in location_horizontal:
    
    for vertical in location_vertical:
        loc = horizontal + '-' + vertical
        loc_list.append(loc)
```

```python
rx1_raw = []
rx2_raw = []
rx3_raw = []
rx1_lin = []
rx2_lin = []
rx3_lin = []
for Rx in receiver:
    
    for horizontal in location_horizontal:
        for vertical in location_vertical:
            
            loc = horizontal + '-' + vertical
   
            
            path = Rx + '-' + horizontal + '-' + vertical
            data_list = ['(rssi)']
            
            for data in data_list:
                filename = path+data+'.txt'
                
                                   
                f = open(filename, 'r')
                line = f.readline()[1:-1]
                rssi = line.split(',')
                #rssi = np.array(rssi).astype(int)
                    
                       
                       
                if Rx == '1':
                    rx1_raw.append(rssi)
                    elist = []
                    for i in rssi:
                        k = 10 ** (-int(i)/10)
                        elist.append(k)
                    rx1_lin.append(elist)
                
                elif Rx == '2':
                    rx2_raw.append(rssi)
                    elist = []
                    for i in rssi:
                        k = 10 ** (-int(i)/10)
                        elist.append(k)
                    rx2_lin.append(elist)
                
                elif Rx == '3':
                    rx3_raw.append(rssi)
                    elist = []
                    for i in rssi:
                        k = 10 ** (-int(i)/10)
                        elist.append(k)
                    rx3_lin.append(elist)
```

```python
def remove_values_from_list(the_list, val):
    while val in the_list:
        the_list.remove(val)
    return np.array(the_list).astype(int)
```

```python
rx1 = []
rx2 = []
rx3 = []
for number in range(3):


    for i in range(20):
        race_for_out = pd.DataFrame(rx1_raw[i])
        clf = IsolationForest(max_samples = 1000, random_state = 1)
        clf.fit(race_for_out)
        y_pred_outliers = clf.predict(race_for_out)

        index = 0
        for j in y_pred_outliers:
            if(j==-1):
                rx1_raw[i][index] = 0
            index = index+1
        rx1.append(remove_values_from_list(rx1_raw[i], 0))

    for i in range(20):
        race_for_out = pd.DataFrame(rx2_raw[i])
        clf = IsolationForest(max_samples = 1000, random_state = 1)
        clf.fit(race_for_out)
        y_pred_outliers = clf.predict(race_for_out)

        index = 0
        for j in y_pred_outliers:
            if(j==-1):
                rx2_raw[i][index] = 0
            index = index+1
        rx2.append(remove_values_from_list(rx2_raw[i], 0))

    for i in range(20):
        race_for_out = pd.DataFrame(rx3_raw[i])
        clf = IsolationForest(max_samples = 1000, random_state = 1)
        clf.fit(race_for_out)
        y_pred_outliers = clf.predict(race_for_out)

        index = 0
        for j in y_pred_outliers:
            if(j==-1):
                rx3_raw[i][index] = 0
            index = index+1
        rx3.append(remove_values_from_list(rx3_raw[i], 0))
```

## data augmentation

```python
#dbm   


np.random.seed(100)
n = 10000
rx1_aug = []
rx2_aug = []
rx3_aug = []

for i in rx1:
    Data = np.round( np.random.normal(np.mean(np.array(i)), np.std(np.array(i)), n) , 5 )
    rx1_aug.append(Data)

for i in rx2:
    Data = np.round( np.random.normal(np.mean(np.array(i)), np.std(np.array(i)), n) , 5 )
    rx2_aug.append(Data)
    
for i in rx3:
    Data = np.round( np.random.normal(np.mean(np.array(i)), np.std(np.array(i)), n) , 5 )
    rx3_aug.append(Data)
```

```python
#

rx1_aug_lin = np.zeros([20,n])
rx2_aug_lin = np.zeros([20,n])
rx3_aug_lin = np.zeros([20,n])
k = 0.00001

#0.00001  + dbm 60%
#0.000001 + dbm 60%
#1   - dbm 30%
#10000 -dbm 60%

for i in range(20):
    for j in range(n):
        rx1_aug_lin[i][j] =(10 ** (-rx1_aug[i][j]/10)) *k
        
for i in range(20):
    for j in range(n):
        rx2_aug_lin[i][j] =(10 ** (-rx2_aug[i][j]/10)) *k
        
for i in range(20):
    for j in range(n):
        rx3_aug_lin[i][j] =(10 ** (-rx3_aug[i][j]/10)) *k
```

## x_train, y_train 

```python
x_train = []
for i in range(20):
    for j in range(n):
        #value = [rx1_aug[i][j], rx2_aug[i][j], rx3_aug[i][j]]
        value = [rx1_aug_lin[i][j], rx2_aug_lin[i][j], rx3_aug_lin[i][j]]
        #value = [rx1[i][j], rx2[i][j], rx3[i][j]]
        x_train.append(value)
np.array(x_train).shape
```

```python
def labeling(label, class_num = 20):
    label_ = np.zeros(class_num).astype(int)
    label_[label - 1] = 1
    return label_.tolist()
```

```python
y_train = []
for i in range(20):
    for j in range(n):
        y_train.append(labeling(i+1))
np.array(y_train).shape
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
combine = list(zip(x_train, y_train))

random.shuffle(combine)

x_train, y_train = zip(*combine)

x_train = np.array(x_train)

y_train = np.array(y_train)
```

## Model 

```python
from keras import models
from keras import layers
from keras.layers import Dropout
from keras.optimizers import Adam

%matplotlib inline
import matplotlib.pyplot as plt

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64, activation='elu',
                           input_shape=(3,)))
    model.add(Dropout(0.3))
    model.add(layers.Dense(64, activation='elu'))
    model.add(Dropout(0.3))
    model.add(layers.Dense(20, activation = 'softmax'))
    model.compile(optimizer = 'Adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = build_model()
model.summary()
```

```python
hist = model.fit(x_train, y_train, epochs=30, batch_size=64, validation_split=0.2)
```
