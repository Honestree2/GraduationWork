## 기본변수설정

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
receiver = ['1','2','3']
path_lists = ['1-1.5', '2-2.5', '3-3.5', '4-2.5', '5-3.5']
ch_list = list(range(1,14))
```

```python
ch_list = list(range(1,14))
loc_list = []
for path_list in path_lists:
        loc = path_list
        loc_list.append(loc)
        
df_1_test = pd.DataFrame(columns = loc_list, index = ch_list)
df_2_test = pd.DataFrame(columns = loc_list, index = ch_list)
df_3_test = pd.DataFrame(columns = loc_list, index = ch_list)
```

```python
for Rx in receiver:
    
    print('')
    print('수신기 : ', Rx)
    print('')
    for path_list in path_lists:
        loc = path_list
        path = Rx + '-'+path_list
        data_list = ['(channel)','(rssi)']
            
        for data in data_list:
                
            filename = path+data+'.txt'
                
            if data == '(channel)':
                    
                f = open(filename, 'r')
                line = f.readline()[1:-1]
                channel = line.split(',')
                channel = np.array(channel).astype(int)
                    
                print('채널 리스트 : ',channel)
                print('')
                    
            else:
                    
                f = open(filename, 'r')
                line = f.readline()[1:-1]
                rssi = line.split(',')
                rssi = np.array(rssi).astype(int)
                    
                print('Rssi 리스트 : ',rssi)
                print('')
                
        sorting = [[],[],[],[],[],[],[],[],[],[],[],[],[]]
            
        index = 0
        for ch in channel:
                
            sorting[ch - 1].append(rssi[index])
            index += 1
                
        print(sorting)
            
        if Rx == '1':
            df_1_test[loc] = sorting
                
        elif Rx == '2':
            df_2_test[loc] = sorting
                
        elif Rx == '3':
            df_3_test[loc] = sorting
```

```python
def make_input(list_all): # 수신기, 위치 넣으면 그 위치 나타내는 input data (13,) shape 만들기
    
    x_input = [] # 결과적으로 shape : (1000000, 13)
        
    list_len = [len(list_all[1]),len(list_all[2]),len(list_all[3]),len(list_all[4]),len(list_all[5]),len(list_all[6]),len(list_all[7]),
                len(list_all[8]),len(list_all[9]),len(list_all[10]),len(list_all[11]),len(list_all[12]),len(list_all[13])]
    
    num = min(list_len) # 1000000
    
    for ch in ch_list:
        
#         print('채널 : ', ch)
        
        if ch == 1:

            for i in range(num):

                x_input.append([list_all[ch][i]])

        else:

            for i in range(num):

                x_input[i].append(list_all[ch][i])
                
#         print(x_input[-1])

        
    return np.array(x_input) # shape : (1000000, 13)
```

```python
def labeling(label, class_num = 20):
    
    label_ = np.zeros(class_num).astype(int)
    
    label_[label - 1] = 1
    
    return list(label_)

labeling(3)
```

## x_test 크롤링

```python
x_test = []

for loc in loc_list:
    
    tmp = []
    
    print('위치 : ', loc)
    
    x_test_1 = make_input(df_1_test[loc])
    x_test_2 = make_input(df_2_test[loc])
    x_test_3 = make_input(df_3_test[loc])
    
    print('make_input done!')
    
    list_num = [len(x_test_1),len(x_test_2),len(x_test_3)]
    
    min_val = min(list_num)
        
    for i in range(min_val):
        
        tmp.append([x_test_1[i]])#, x_train_2[i], x_train_3[i]]) # (1000000, 3, 13)
                
    for i in range(min_val):
        
        tmp[i].append(x_test_2[i])
        
    for i in range(min_val):
        
        tmp[i].append(x_test_3[i])
        
    x_test.append(tmp)
        
# x_test shape : (20, x, 3, 13)
```

```python
for i in range(5):
    print(len(x_test[i]))
    
mylist = []
for i in range(5):
    mylist.append(len(x_test[i]))
```
